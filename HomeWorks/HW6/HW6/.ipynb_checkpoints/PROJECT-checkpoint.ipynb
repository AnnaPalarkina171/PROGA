```python
import urllib.request
import json
from datetime import datetime
import sqlite3
import time
import os
import re
from pymystem3 import Mystem
from collections import Counter
from nltk.corpus import stopwords
import operator
import matplotlib
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
mystem = Mystem()

token = '70058a7d0b3a939aa6ec3b0c0d6073d36cf57b3ac1a52563632984ecd4a3cd15db93f530713f2d86b5c4f'

def request(url):
    req = urllib.request.Request(url)
    response = urllib.request.urlopen(req)
    result = response.read().decode('utf-8')
    data = json.loads(result)
    return data

def write(text_for_corpus):
    with open('corpus\plain_texts.txt', 'a', encoding='utf-8') as f:
        f.write(text_for_corpus)
        f.write(' ')

def posts():
    posts_id = []
    conn = sqlite3.connect('database.db')
    c = conn.cursor() 
    c.execute("CREATE TABLE IF NOT EXISTS posts(id, text, lemmas, len, year, month)")
    url = 'https://api.vk.com/method/wall.get?owner_id=-22822305&count=110&v=5.92&access_token=%s' % (token)
    data = request(url)
    text = data['response']['items']
    for x in text:
        post_id = x['id']                              #ID post
        time_post = x['date']
        utc = datetime.fromtimestamp(time_post)
        post_year = str(utc).split('-')[0]             #YEAR post
        post_month = str(utc).split('-')[1]            #MONTH post
        if x['text'] == '':
            dic = x['copy_history']
            post_text = dic[0]['text']                 #TEXT post
            txt = dic[0]['text']
            txt = re.sub(r'\n', r' ', txt)
            txt = re.sub(r'[^а-яА-ЯёЁ ]', '', txt)
            txt = re.sub(r'   ', r' ', txt)
            text_lem = [text for text in mystem.lemmatize(txt)]
            post_lemmatized = ''.join(text_lem)       #LEMMATIZED post
            post_lem = []
            for word in text_lem:
                if word != ' ':
                    post_lem.append(word)
            post_len = len(post_lem)-1               #LEN post
        else:
            post_text = x['text']                    #TEXT post
            txt = re.sub(r'\n', r' ', post_text)
            txt = re.sub(r'[^а-яА-ЯёЁ ]', '', txt)
            txt = re.sub(r'   ', r' ', txt)
            text_lem = [text for text in mystem.lemmatize(txt)]
            post_lemmatized = ''.join(text_lem)       #LEMMATIZED post
            post_lem = []
            for word in text_lem:
                if word != ' ':
                    post_lem.append(word)
            post_len = len(post_lem)-1               #LEN post
            
            lst = []
            lst.append(post_id)
            posts_id.append(post_id)
            lst.append(post_text)
            lst.append(post_lemmatized)
            lst.append(post_len)
            lst.append(post_year)
            lst.append(post_month)
            c.execute('INSERT INTO posts VALUES (?, ?, ?, ?, ?, ?)', lst)
            conn.commit()
            
            text_for_corpus = post_text
            write(text_for_corpus)
    return posts_id
                       
def comments(post_id):
    conn = sqlite3.connect('database.db')
    c = conn.cursor()
    c.execute("CREATE TABLE IF NOT EXISTS comments(post id, comment id, person id, text, lemmas, len, city, sex)")
    for pst_id in posts_id:
        url = 'https://api.vk.com/method/wall.getComments?owner_id=-22822305&post_id=%s&count=110&v=5.92&access_token=%s' % (
        pst_id, token)
        data = request(url)
        text = data['response']['items']
        for x in text:
            if 'deleted' not in x.keys():
                id_c = str(x['from_id'])                               #иногда комментят группы
                if id_c[0] != '-':
                    txt = x['text']
                    txt = re.sub(r'\n', r' ', txt)
                    txt = re.sub(r'[^а-яА-ЯёЁ ]', '', txt)
                    txt = re.sub(r'   ', r' ', txt)
                    if txt != '':
                        com_text = x['text']
                        com_post_id = x['post_id']
                        com_id = x['id']
                        from_id = x['from_id']
                        text_lem = [text for text in mystem.lemmatize(txt)]
                        com_lemmatized = ''.join(text_lem)
                        com_lem = []
                        for word in text_lem:
                            if word != ' ':
                                com_lem.append(word)
                        com_len = len(com_lem)-1

                        #достаем инфо со странички комментатора
                        url = 'https://api.vk.com/method/users.get?v=5.92&access_token=%s&fields=city&user_ids=%s' % (token, com_id)
                        data = request(url)
                        if 'city' not in (data['response'][0].keys()):
                            cities = 'none'
                        else:
                            city = data['response'][0]['city']
                            cities = city['title']

                        url = 'https://api.vk.com/method/users.get?v=5.92&access_token=%s&fields=sex&user_ids=%s' % (token, com_id)
                        data = request(url)
                        if 'sex' not in (data['response'][0].keys()):
                            sex_com = 'none'
                        else:
                            sex = data['response'][0]['sex']
                            if sex == 1:
                                sex_com = 'female'
                            if sex == 2:
                                sex_com = 'male'


                        lst = []
                        lst.append(com_post_id)
                        lst.append(com_id)
                        lst.append(from_id)
                        lst.append(com_text)
                        lst.append(com_lemmatized)
                        lst.append(com_len)
                        lst.append(cities)
                        lst.append(sex_com)
                        c.execute('INSERT INTO comments VALUES (?, ?, ?, ?, ?, ?, ?, ?)', lst)
                        conn.commit()

                        text_for_corpus = com_text
                        write(text_for_corpus)
                        

def lemmatizer():
    path = path = os.getcwd()
    plain = path + '\\corpus\\plain_texts.txt'
    mystem = path + '\\corpus\\mystem_texts.xml'
    os.system(r'C:\\Users\\User\\Desktop\\mystem.exe -cdgin --format xml --eng-gr {} {}'.format(plain, mystem))
    mystem_xml = path + '\\corpus\\mystem_texts.xml'
    with open(mystem_xml, 'r', encoding='utf-8') as f:
        text = f.read()
        list_words = re.findall(r'lex="(.*?)"', text)
        with open('corpus\lemmatized_texts.txt', 'w', encoding='utf-8') as f:
            f.write(' '.join(list_words))
            
            
#posts_id = posts() 
#comments(posts_id)
#lemmatizer()


 
```
